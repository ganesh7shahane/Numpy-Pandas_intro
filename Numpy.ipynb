### Removing duplicate entries from a Numpy arrays and comparing them to find unique elements amongst them

```python
ls -ltrh *.txt
```

```python
import numpy as np
from functools import reduce

core = np.loadtxt('space_core_waters.txt')
print(core.size)
upper = np.loadtxt('upper_waters.txt')
print(upper.size)
lower = np.loadtxt('lower_waters.txt')
print(lower.size)

c_unique = np.unique(core) #find unique elements
print(c_unique.size)
u_unique = np.unique(upper)
print(u_unique.size)

l_unique = np.unique(lower)
print(l_unique.size)

common_values = reduce(np.intersect1d, ( u_unique, l_unique, c_unique)) #finding common resid amongst the 3 arrays

print(common_values.size)
```
